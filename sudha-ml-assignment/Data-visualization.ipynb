## Data Visualization in Statistics
It is the graphical representation of the information and data. It helps us see and understand trends, patterns and outliers in data by using visualization tools and visual elements like graphs, charts, and maps. It is essential to analyse massive amoount of information and make data-driven decisions. It is another form of visual art to present data to non-technical audiences without confusion.

It helps people see, interact with and better understand data. Every STEM field benefits from understanding data - and so do fields in government, finance, marketing, history, education, sports etc. In short, it is analysis and virtual story telling using data.

## General types of visualizations

Charts: Graph, diagram or map
Table: Rows and column
Graph: Points, lines, segments, curves, or areas that represent certain variables in comparison to each other.
Geospatial: Shows data in map form using different shapes and colors to show relationship between pieces of data and specific locations.
Infographic: Combination of visuals and words that represent data. Usually charts and diagrams.
Dashboard: A collection of visualizations and data displayed in one place

## Specific types of visualizations:

### Bar chart
Bar charts represent numerical values compared to each other. The length of the bar represents the value of each variable.


```python
#represnting bar chart using matplotlib
import matplotlib.pyplot as plt
   
Country = ['USA','Canada','Germany','UK','France']
GDP_Per_Capita = [45000,42000,52000,49000,47000]

plt.bar(Country, GDP_Per_Capita)
plt.title('Country Vs GDP Per Capita')
plt.xlabel('Country')
plt.ylabel('GDP Per Capita')
plt.show()
```

### Box-and-whisker plot
It is a simple way of representing statistical data on a plot in which a rectangle is drawn to represent the second and third quartiles, usually with a vertical line inside to indicate the median value. The lower and upper quartiles are shown as horizontal lines either side of the rectangle. The whiskers are the two lines outside the box, that go from the minimum to the lower quartile (the start of the box) and then from the upper quartile (the end of the box) to the maximum. 

```python
#representing box-and-whisker plot using matplotlib in a dataset
import pandas as pd
data=pd.read_csv("women-clothing-dataset.csv")
data.head()
```

```python
data.drop('Unnamed: 0', inplace=True, axis=1)
```

```python
#creating the box whisker plot of single attribute
plt.rcParams['figure.figsize']=10,5
data.boxplot('Clothing ID')
```

```python
#creating a bo-whisker plot of entire data set
data.boxplot()
```

### Histograms
A histogram is a graphical representation of data points organized into user-specified ranges. Similar in appearance to a bar graph, the histogram condenses a data series into an easily interpreted visual by taking many data points and grouping them into logical ranges or bins.

```python
data.hist(bins=30, figsize=(20,30))
```
