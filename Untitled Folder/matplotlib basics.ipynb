```python
from matplotlib import pyplot as plt
import numpy as np
np.random.seed(4)

x = np.linspace(0,20,21,dtype=np.int32)
# s = x /2 + np.random.random((21)-np.random.random(21))

y = np.linspace(0,10,21)

plt.figure(figsize=(15,8))
plt.title("Example graph")

plt.xlabel("Volage")
plt.ylabel("Temperature")

s_plot = plt.scatter(x,y, label = "observed value")

y_plot = plt.plot(y,label="fit y")
plt.legend()
plt.show()
```

```python
category = ['a','b','c','d','e']
freq = [5,2,7,8,2]
plt.bar(category,freq,label="sth")
plt.legend()
plt.xlabel("sajk")
plt.ylabel("dsa")
plt.title("uweywq")
plt.show()
```
